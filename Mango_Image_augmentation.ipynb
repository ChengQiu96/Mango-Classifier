```python
import cv2
import numpy as np
from skimage import io 
from skimage.transform import rotate, AffineTransform, warp
import matplotlib.pyplot as plt
import random
from skimage import img_as_ubyte
import os
from skimage.util import random_noise
from tqdm import tqdm
```

```python
def anticlockwise_rotation(image):
    angle= random.randint(0,180)
    return rotate(image, angle)

def clockwise_rotation(image):
    angle= random.randint(0,180)
    return rotate(image, -angle)

def h_flip(image):
    return  np.fliplr(image)

def v_flip(image):
    return np.flipud(image)

def add_noise(image):
    return random_noise(image)

def blur_image(image):
    return cv2.GaussianBlur(image, (9,9),0)
```

```python
images_path ="Mango"
augmented_path ="Aug_Mango_2"
images = []
    

for im in os.listdir(images_path):  # read image name from folder and append its path into "images" array     
    images.append(os.path.join(images_path,im))

i=1       
for x in images:    
    image = x
    original_image = io.imread(image)
    small_image = cv2.resize(original_image,(200, 200))
    angle= random.randint(0,180)
    counterclock_image = rotate(small_image,angle)
    clock_image = rotate(small_image,-angle)
    h_flip_image = np.fliplr(small_image)
    v_flip_image = np.flipud(small_image)
    noise_image = random_noise(small_image)
    blur_image = cv2.GaussianBlur(small_image, (9,9),0)
    
    new_image_path1= "%s/small_%s.jpg" %(augmented_path, i)
    new_image_path2= "%s/counterclock_%s.jpg" %(augmented_path, i)
    new_image_path3= "%s/clock_%s.jpg" %(augmented_path, i)
    new_image_path4= "%s/h_flip_%s.jpg" %(augmented_path, i)
    new_image_path5= "%s/v_flip_%s.jpg" %(augmented_path, i)
    new_image_path6= "%s/noise_%s.jpg" %(augmented_path, i)
    new_image_path7= "%s/blur_%s.jpg" %(augmented_path, i)
    
    
    transformed_image1 = img_as_ubyte(small_image)#Convert an image to unsigned byte format, with values in [0, 255].
    transformed_image2 = img_as_ubyte(counterclock_image)
    transformed_image3 = img_as_ubyte(clock_image)
    transformed_image4 = img_as_ubyte(h_flip_image)
    transformed_image5 = img_as_ubyte(v_flip_image)
    transformed_image6 = img_as_ubyte(noise_image)
    transformed_image7 = img_as_ubyte(blur_image)
    
    
    transformed_image1=cv2.cvtColor(transformed_image1, cv2.COLOR_BGR2RGB) #convert image to RGB before saving it
    transformed_image2=cv2.cvtColor(transformed_image2, cv2.COLOR_BGR2RGB)
    transformed_image3=cv2.cvtColor(transformed_image3, cv2.COLOR_BGR2RGB)
    transformed_image4=cv2.cvtColor(transformed_image4, cv2.COLOR_BGR2RGB)
    transformed_image5=cv2.cvtColor(transformed_image5, cv2.COLOR_BGR2RGB)
    transformed_image6=cv2.cvtColor(transformed_image6, cv2.COLOR_BGR2RGB)
    transformed_image7=cv2.cvtColor(transformed_image7, cv2.COLOR_BGR2RGB)
    
    cv2.imwrite(new_image_path1, transformed_image1) # save transformed image to path
    cv2.imwrite(new_image_path2, transformed_image2)
    cv2.imwrite(new_image_path3, transformed_image3)
    cv2.imwrite(new_image_path4, transformed_image4)
    cv2.imwrite(new_image_path5, transformed_image5)
    cv2.imwrite(new_image_path6, transformed_image6)
    cv2.imwrite(new_image_path7, transformed_image7)
    
    
    
    
    i =i+1
    
#to generate more images, put above 3 statement inside while n<... loop
```
